```python
import csv
import googletrans

row = []
# Import the CSV file
with open("af.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")

    # Create a translator object
    translator = googletrans.Translator()

    # Iterate over the rows in the CSV file
    for i,row in enumerate(reader):
        # Get the English word from the first column
        english_word = row[0]

        # Translate the English word to Mongolian
        mongolian_translation = translator.translate(english_word, dest="mn")

        # Add the Mongolian translation to the second column
        row[1] =(mongolian_translation.text)
        print(i/9905)
        print(row)

        # Write the updated CSV file
        with open("new-frappe.csv", "a", newline="") as csvfile:
            writer = csv.writer(csvfile, delimiter=",", escapechar="\\")
            writer.writerow(row)
```
